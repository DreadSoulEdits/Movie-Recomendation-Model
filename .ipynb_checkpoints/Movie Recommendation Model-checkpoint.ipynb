# Importing Dependencies

```python
import numpy as np
import pandas as pd
import seaborn as sns
import difflib
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# Collecting Data

```python
movieData = pd.read_csv('movies.csv')
```

```python
movieData.head()
```

```python
movieData.shape
```

# Feature Selection

```python
selectedFeatures = ['genres', 'keywords', 'overview', 'cast', 'director']
```

```python
for feature in selectedFeatures:
    movieData[feature] = movieData[feature].fillna('')
```

```python
# combined features

combinedFeatures = movieData['genres']+ movieData['keywords']+ movieData['overview']+ movieData['cast']+ movieData['director']
```

```python
print(combinedFeatures)
```

```python
vectorizer = TfidfVectorizer()
```

```python
featureVectors = vectorizer.fit_transform(combinedFeatures)
```

```python
print(featureVectors)
```

# Cosine Similarity

```python
# getting the similarity score using cosine similarity
similarity = cosine_similarity(featureVectors)
```

```python
print(similarity)
print(similarity.shape)
```

```python
# getting Movie name from user

movieName = input("Enter your Fav movie name: ")
```

```python
titleList = movieData['title'].tolist()
print(titleList)
```

```python
# finding the close match for the movie name
findCloseMatch = difflib.get_close_matches(movieName, titleList)
print(findCloseMatch)
```

```python
closeMatch = findCloseMatch[0]
print(closeMatch)
```

```python
# finding the index of the movie with title
movieIndex = movieData[movieData.title == closeMatch]['index'].values[0]
print(movieIndex)
```

```python
# getting list of similar movies
similarityScore = list(enumerate(similarity[movieIndex]))
print(similarityScore)
```

```python
sortedSimilarMovies = sorted(similarityScore, key = lambda x:x[1], reverse = True)
print(sortedSimilarMovies)
```

```python
# printing the names of the similar movies
print('Movies suggested for you: ')

i = 1

for movie in sortedSimilarMovies:
    index = movie[0]
    movieTitle = movieData[movieData.index == index]['title'].values[0]
    if (i < 11):
        print(i, '.', movieTitle)
        i+=1
```
